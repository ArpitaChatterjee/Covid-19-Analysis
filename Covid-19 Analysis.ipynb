# Q2

```python
df.drop(['Date', 'Time', 'ConfirmedIndianNational', 'ConfirmedForeignNational', 'Unnamed: 9'], axis=1, inplace=True)
```

```python
df.columns
```

# Q3

```python
data = df.groupby('State/UnionTerritory').sum()
data.reset_index(inplace=True)
data.sort_values('Confirmed', inplace=True, ascending=False)
data
```

```python
data["Active"]= (data["Confirmed"]- data["Cured"]-data["Deaths"])
data.columns
```

# Q6

```python
data['ratio']=(data['Deaths']/data['Cured'])
data.columns
```

```python
data
```

```python
data.sort_values('ratio', inplace=True, ascending=False)
data
```

```python
#sns.set_theme(style='whitegrid')
fig, ax = plt.subplots(figsize = (12,6))
sns.barplot(x=data['State/UnionTerritory'], y=data["ratio"], data=data, label="Ratio", ax=ax, ci=None)
ind = np.arange(len(data['State/UnionTerritory']))
ax.set_xticks(range(0,len(ind)))
ax.set_xticklabels(list(data['State/UnionTerritory']), rotation= 90)
```
